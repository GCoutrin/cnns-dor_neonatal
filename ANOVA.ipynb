# Introdução

Este notebook apresenta a aplicação da Análise de Variância (ANOVA) para a comparação quantitativa do desempenho de 5 CNNs

# Preparação dos dados

Carregando os resultados dos modelos

```python
import pandas as pd
import numpy as np
from scipy.stats import f_oneway, bartlett, shapiro
import os
import os.path as osp

path = 'caminho dos arquivos de resultados'

files = sorted(os.listdir(path))
print(files)
```

```python
# resultados por modelo
vgg = pd.read_excel(osp.join(path,files[-1]),index_col=0).drop(columns=['final'])
incep = pd.read_excel(osp.join(path,files[2]),index_col=0).drop(columns=['final'])
resnet = pd.read_excel(osp.join(path,files[5]),index_col=0).drop(columns=['final'])
senet = pd.read_excel(osp.join(path,files[-2]),index_col=0).drop(columns=['final'])
ncnn = pd.read_excel(osp.join(path,files[3]),index_col=0).drop(columns=['final'])
```

```python
# coletando as métricas de desempenho

vgg_acc = vgg.loc['Acurácia'].values
vgg_f1 = vgg.loc['Pontuação F1'].values
vgg_AUC = vgg.loc['AUC'].values

resnet_acc =resnet.loc['Acurácia'].values
resnet_f1 = resnet.loc['Pontuação F1'].values
resnet_AUC =resnet.loc['AUC'].values

senet_acc =senet.loc['Acurácia'].values
senet_f1 = senet.loc['Pontuação F1'].values
senet_AUC =senet.loc['AUC'].values

incep_acc =incep.loc['Acurácia'].values
incep_f1 = incep.loc['Pontuação F1'].values
incep_AUC =incep.loc['AUC'].values

ncnn_acc =ncnn.loc['Acurácia'].values
ncnn_f1 = ncnn.loc['Pontuação F1'].values
ncnn_AUC =ncnn.loc['AUC'].values
```

# Suposições para realização do ANOVA

## Verificação da distribuição Normal - Shapiro Wilk

```python
# listas auxiliares
shapiro_acc =[]
shapiro_f1  =[]
shapiro_auc =[]
```

```python
#VGG
shapiro_test_ACC = shapiro(vgg_acc)
shapiro_test_F1 = shapiro(vgg_f1)
shapiro_test_AUC = shapiro(vgg_AUC)
print("ACC: ",shapiro_test_ACC)
print("F1: ",shapiro_test_F1)
print("AUC: ",shapiro_test_AUC)

shapiro_acc.append(round(shapiro_test_ACC.pvalue,2))
shapiro_f1.append(round(shapiro_test_F1.pvalue,2))
shapiro_auc.append(round(shapiro_test_AUC.pvalue,2))
```

```python
#ResNet 

shapiro_test_ACC = shapiro(resnet_acc)
shapiro_test_F1 = shapiro(resnet_f1)
shapiro_test_AUC = shapiro(resnet_AUC)
print("ACC: ",shapiro_test_ACC)
print("F1: ",shapiro_test_F1)
print("AUC: ",shapiro_test_AUC)

shapiro_acc.append(round(shapiro_test_ACC.pvalue,2))
shapiro_f1.append(round(shapiro_test_F1.pvalue,2))
shapiro_auc.append(round(shapiro_test_AUC.pvalue,2))
```

```python
#SENet

shapiro_test_ACC = shapiro(senet_acc)
shapiro_test_F1 = shapiro(senet_f1)
shapiro_test_AUC = shapiro(senet_AUC)
print("ACC: ",shapiro_test_ACC)
print("F1: ",shapiro_test_F1)
print("AUC: ",shapiro_test_AUC)

shapiro_acc.append(round(shapiro_test_ACC.pvalue,2))
shapiro_f1.append(round(shapiro_test_F1.pvalue,2))
shapiro_auc.append(round(shapiro_test_AUC.pvalue,2))
```

```python
#Incep

shapiro_test_ACC = shapiro(incep_acc)
shapiro_test_F1 = shapiro(incep_f1)
shapiro_test_AUC = shapiro(incep_AUC)
print("ACC: ",shapiro_test_ACC)
print("F1: ",shapiro_test_F1)
print("AUC: ",shapiro_test_AUC)

shapiro_acc.append(round(shapiro_test_ACC.pvalue,2))
shapiro_f1.append(round(shapiro_test_F1.pvalue,2))
shapiro_auc.append(round(shapiro_test_AUC.pvalue,2))
```

```python
#NCNN

shapiro_test_ACC = shapiro(ncnn_acc)
shapiro_test_F1 = shapiro(ncnn_f1)
shapiro_test_AUC = shapiro(ncnn_AUC)
print("ACC: ",shapiro_test_ACC)
print("F1: ",shapiro_test_F1)
print("AUC: ",shapiro_test_AUC)

shapiro_acc.append(round(shapiro_test_ACC.pvalue,2))
shapiro_f1.append(round(shapiro_test_F1.pvalue,2))
shapiro_auc.append(round(shapiro_test_AUC.pvalue,2))
```

```python
shapiro_df = pd.DataFrame(data=[shapiro_acc,shapiro_f1,shapiro_auc],columns=['VGG-16','ResNet50','SENet50','Inception-V3','N-CNN'],index=['Acurácia','Pontuação F1','AUC'])
display(shapiro_df)
```

## Homogeneidade das variâncias - teste de Bartlett

```python
# ACC
bartlett_list = []

bartlett_test = bartlett(vgg_acc, incep_acc, resnet_acc, senet_acc, ncnn_acc)
print("ACC: ",bartlett_test)
bartlett_list.append(str(round(bartlett_test.pvalue,2)).replace(".",","))
```

```python
# F1

bartlett_test = bartlett(vgg_f1, incep_f1, resnet_f1, senet_f1, ncnn_f1)
print("F1: ",bartlett_test)
bartlett_list.append(str(round(bartlett_test.pvalue,2)).replace(".",","))
```

```python
# AUC

bartlett_test = bartlett(vgg_AUC, incep_AUC, resnet_AUC, senet_AUC, ncnn_AUC)
print("AUC: ",bartlett_test)
bartlett_list.append(str(round(bartlett_test.pvalue,2)).replace(".",","))
```

```python
bar_df = pd.DataFrame(data=bartlett_list, index = ['Acurácia','Pontuação F1','AUC'], columns=['p-valor'])
display(bar_df)
```

# ANOVA

## ANOVA com todos os modelos

```python
# Acuracia

anova_test = f_oneway(vgg_acc, incep_acc, resnet_acc, senet_acc, ncnn_acc)
print("ACC: ",anova_test)
```

```python
# F1

anova_test = f_oneway(vgg_f1, incep_f1, resnet_f1, senet_f1, ncnn_f1)
print("F1: ",anova_test)
```

```python
# AUC

anova_test = f_oneway(vgg_AUC, incep_AUC, resnet_AUC, senet_AUC, ncnn_AUC)
print("AUC: ",anova_test)
```

## ANOVA com todos os modelos, exceto a N-CNN

```python
# Acuracia

anova_test = f_oneway(vgg_acc, incep_acc, resnet_acc, senet_acc)
print("ACC: ",anova_test)
```

```python
# F1

anova_test = f_oneway(vgg_f1, incep_f1, resnet_f1, senet_f1)
print("F1: ",anova_test)
```

```python
# AUC

anova_test = f_oneway(vgg_AUC, incep_AUC, resnet_AUC, senet_AUC)
print("AUC: ",anova_test)
```
